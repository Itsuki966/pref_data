```python
import os
import pickle
import pandas as pd
```

## データの読み込み

```python
def data_load(path):
    data = pd.read_csv(path, header=0)
    return data
```

```python
dir_path = "/Users/itsukikuwahara/Desktop/codes/research/data/data_pref/"

# パスで指定したディレクトリ内のファイルとディレクトリを取得
files = os.listdir(dir_path)

# ファイルとディレクトリを分ける
files_file = [f for f in files if os.path.isfile(os.path.join(dir_path, f))]
files_dir = [f for f in files if os.path.isdir(os.path.join(dir_path, f))]

# .DS_Storeを削除
files_file = [f for f in files_file if f != ".DS_Store"]
files_dir = [f for f in files_dir if f != ".DS_Store"]

# ファイルとディレクトリを確認
print(f"file : {files_file}", f"dir : {files_dir}")

# データを保存するための辞書を用意
df_dict = {}

# file（ディレクトリのないカテゴリ）を辞書に追加
for f in files_file:
    temp = []
    file_path = dir_path + f
    df = data_load(file_path)
    temp.append(df)
    df_dict[f.split(".")[0]] = temp
    
    
# ディレクトリがあるカテゴリを辞書に追加する処理
for d in files_dir:
    # ディレクトリのパスを取得
    category = d
    files_dir_files = os.listdir(dir_path + category)
    
    # .DS_Storeを削除
    files_dir_files = [f for f in files_dir_files if f != ".DS_Store"]

    # カテゴリごとのデータを格納するリストを用意
    df_list = []
    
    # ディレクトリ内のファイルを読み込んでリストに追加
    for ff in files_dir_files:
        file_path = dir_path + category + "/" + ff
        df = data_load(file_path)
        df_list.append(df)
    
    # カテゴリごとのデータを辞書に追加
    df_dict[category] = df_list
```

## 若年人口

```python
def young_population(df):
    df["若年人口"] = df["A1201_0～4歳人口【人】"] + df["A1202_5～9歳人口【人】"] + df["A1203_10～14歳人口【人】"] + df["A1204_15～19歳人口【人】"] + df["A1205_20～24歳人口【人】"] + df["A1206_25～29歳人口【人】"] + df["A1207_30～34歳人口【人】"]
    return df
```

```python
for i in range(len(df_dict["population"])):
    df_dict["population"][i] = young_population(df_dict["population"][i])
```

## 説明変数を処理

### 各項目のデータフレームを結合する

```python
def merge_df(base_df, df_dictionary):
    for key in df_dictionary.keys():
        if key != "population":
            for i in range(len(df_dictionary[key])):
                df = df_dictionary[key][i]
                base_df = pd.merge(base_df, df, on=["調査年","地域"], how="left")
        else:
            pass
    return base_df           
```

```python
merged_df = merge_df(df_population, df_dict)

merged_df.drop(["/Ａ　人口・世帯","/Ｂ　自然環境","/Ｄ　行政基盤","/Ｊ　福祉・社会保障","/Ｉ　健康・医療","/Ｃ　経済基盤","/Ｅ　教育","/Ｈ　居住","/Ｇ　文化・スポーツ","/Ｆ　労働"], inplace=True, axis=1)
```

```python
# 各項目の欠損値の数の確認
temp = dict(merged_df.isna().sum())
temp_dict = {}

for key in temp.keys():
    if temp[key] != 0:
        print(key, temp[key])
        temp_dict[key] = temp[key]
```

### 年度ごとにデータフレームを作成する

```python
object_df = merged_df
year_list = ['1975年度', '1976年度', '1977年度', '1978年度', '1979年度', '1980年度', '1981年度', '1982年度', '1983年度', '1984年度', '1985年度', '1986年度', '1987年度', '1988年度', '1989年度', '1990年度', '1991年度', '1992年度', '1993年度', '1994年度', '1995年度', '1996年度', '1997年度', '1998年度', '1999年度', '2000年度', '2001年度', '2002年度', '2003年度', '2004年度', '2005年度', '2006年度', '2007年度', '2008年度']#, '2009年度', '2010年度', '2011年度', '2012年度', '2013年度', '2014年度', '2015年度', '2016年度', '2017年度', '2018年度', '2019年度', '2020年度']
year_df_dict = {}

for y in year_list:
    year = int(y.split("年")[0])
    year_df_dict[year] = object_df[object_df["調査年"]==y]
```

```python
key_list = []
object_df_dict = {}

# year_listの中の年度のデータが揃っていないデータ項目を抽出
for key in year_df_dict.keys():
    df = year_df_dict[key]
    temp = dict(df.isna().sum())
    for k in temp.keys():
        if temp[k] != 0:
            key_list.append(k)

# データが揃っていないデータ項目を削除
del_list = set(key_list)

for key in year_df_dict.keys():
    df = year_df_dict[key]
    df_new = df.drop(del_list, axis=1)    
    object_df_dict[key] = df_new
```
